```r
n=300
```

```r
sigma_eta=0.5
```

```r
eta=sigma_eta*rnorm(n)
```

```r
mean(eta)
```

```r
var(eta)
```

```r
ts.plot(eta)
```

```r
acf(eta, drop.lag.0=FALSE, lag.max = 120 )
```

```r
sigma_e=1
e=sigma_e*rnorm(n)
```

```r
ts.plot(e)
```

```r
acf(e,lag.max = 80)
```

```r
q=sigma_eta^2/sigma_e^2
```

```r
mu=0
mu[1]<-1
y=0
```

```r
for(t in 1:(n-1)){
  mu[t+1]=mu[t]+eta[t]
  y[t+1]=mu[t+1]+e[t+1]
}
```

```r
ts.plot(y)
lines(mu,col=45)
```

```r
acf(y,lag.max = 100, drop.lag.0=FALSE)
```

```r
pacf(y,lag.max = 100, drop.lag.0=FALSE)
```

```r
n=400
sigma_eta=0.2
eta=sigma_eta*rnorm(n)
```

```r
sigma_e=0.8
e=sigma_e*rnorm(n)
```

```r
sigma.xi=2.1
```

```r
xi=sigma.xi*rnorm(n)
```

```r
mu=0
mu[1]<-0
beta=0
beta[1]<-0
y=0
for(t in 1:(n-1)){
  beta[t+1]=beta[t]+xi[t]
  mu[t+1]=beta[t]+mu[t]+eta[t]
  y[t+1]=mu[t+1]+e[t+1]
}
```

```r
ts.plot(y)
lines(mu,col="red")
lines(beta,col="blue")
```
